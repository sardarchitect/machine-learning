```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
filepaths = []
for file in os.listdir(f"./data/cadsvskbdp4b4a5njoeh4lp314@group.calendar.google.com/"):
    file = f"./data/cadsvskbdp4b4a5njoeh4lp314@group.calendar.google.com//{file}"
    filepaths.append(file)
```

```python
eventsDict = {}
for file in filepaths:
    if "primary" in file:
        category = "personal"
    if "cadsvskbdp4b4a5njoeh4lp314" in file:
        category = "project"
    if "0eumcrv5gkahhq8o98uej4ame8" in file:
        category = "wellness"
    f = open(file, 'r')
    f_json = json.loads(f.read())
    f_json_events = (f_json['items'])
    if f_json_events != []:
#         print(file)
        for event in f_json_events:
            eventsDict[event['id']] = [category, event['summary'], event['start'], event['end']]
    f.close()
    
## ONLY RUN ONCE
# Some events are full day/multiday events and do not have datetime, this deals with that
for event in eventsDict:
    for i in range(2,4):
        eventDateTime = eventsDict[event][i]
        for key in eventDateTime:
            if 'dateTime' in eventDateTime.keys():
                eventsDict[event][i] = eventDateTime['dateTime']
            if 'date' in eventDateTime.keys():
                eventsDict[event][i] = eventDateTime['date']
```

```python
len(eventsDict)
```

```python
df = pd.DataFrame.from_dict(eventsDict, orient='index')
df.rename({0: 'category', 1:'summary', 2:'start',3:'end'}, axis=1, inplace=True)
df.head()
```

```python
# Convert start and end to datetime objects and deal with timezones
df['start'] = pd.to_datetime(df['start'], utc=True)
df['start'] = df['start'].dt.tz_convert('US/Eastern')
df['end'] = pd.to_datetime(df['end'], utc=True)
df['end'] = df['end'].dt.tz_convert('US/Eastern')

# Create a new column for timedelta durations
df['duration'] = (df['end'] - df['start']).astype('timedelta64[m]')/60
# Create a new column for start day
df['start_date'] = df['start'].dt.date

# Sort by start time
df.sort_values('start', inplace=True)
df.set_index('start_date', inplace=True)
df.drop(columns=['category', 'start', 'end'], inplace=True)
```

## Analysis

1. Total number of hours worked each day
2. Total Weekly Hours worked each week
2. Total number of hours worked on each category

```python
personal_df = df.copy()
work_df = df.copy()
```

```python
personal_df = personal_df[~personal_df['summary'].str.contains('Walker')]
work_df = work_df[work_df['summary'].str.contains('Walker')]
```

```python
# new_df.reset_index(inplace=True)
personal_df = personal_df[personal_df.index < pd.to_datetime('2023-02-01').date()]
work_df = work_df[work_df.index < pd.to_datetime('2023-02-01').date()]
personal_df.head()
```

```python
personal_df = personal_df.groupby(by=personal_df.index).sum()
work_df = work_df.groupby(by=work_df.index).sum()
```

```python
index = pd.date_range('2020-12-10', '2023-01-30')
personal_df = personal_df.reindex(index, fill_value=0)
work_df = work_df.reindex(index, fill_value=0)
```

```python
personal_df.head()
```

```python
print(personal_df['duration'].mean())
print(work_df['duration'].mean())
```

```python
rolling_window = 14

plt.figure(figsize=(24,8))

# Birthday
plt.axvline(dt.datetime(2021,8,14),  color='grey',linestyle='dashed')
plt.axvline(dt.datetime(2022,8,14),  color='grey',linestyle='dashed')
# Out of Country
plt.axvline(dt.datetime(2021,7,11),  color='r',linestyle='dashed')
plt.axvline(dt.datetime(2021,7,31),  color='r',linestyle='dashed')
plt.axvline(dt.datetime(2022,6,4),  color='r',linestyle='dashed')
plt.axvline(dt.datetime(2022,6,25),  color='r',linestyle='dashed')

# Moving
plt.axvline(dt.datetime(2021,5,14),  color='y',linestyle='dashed')
plt.axvline(dt.datetime(2021,6,1),  color='y',linestyle='dashed')
plt.axvline(dt.datetime(2022,7,1),  color='y',linestyle='dashed')
plt.axvline(dt.datetime(2022,7,18),  color='y',linestyle='dashed')
plt.axhline(2,  color='b',linestyle='dashed')

# Background
plt.xticks(rotation=0)
plt.xlabel('Date')

plt.ylim(0, 8)
plt.ylabel('Number of Hours')

plt.grid(color='grey', linestyle='dashed', linewidth=0.5)

plt.plot(personal_df['duration'].rolling(rolling_window).mean())
plt.plot(work_df['duration'].rolling(rolling_window).mean())
plt.savefig('output.png', bbox='tight')

plt.show()
```

```python
personal_df = df.copy()
personal_df = personal_df[~personal_df['summary'].str.contains('Walker')]
personal_df = personal_df[personal_df.index < pd.to_datetime('2023-02-01').date()]
personal_df = personal_df.groupby(by=personal_df.index).sum()
personal_df.index = pd.DatetimeIndex(personal_df.index)
```
