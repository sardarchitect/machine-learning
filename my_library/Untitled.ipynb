```python
# GENERATE RANDOM DATA FROM GAUSSIAN DIST
X = np.random.randn(10000)
# CONVERT DATA INTO PDF
data = utilities.pdf(X, 1)
X = np.array(list(data.keys()))
Y = np.array(list(data.values()))
plt.plot(X,Y)
```

```python
def linear_network(x):
    x = nn.linear(x, x.shape[0], 3)
    y = nn.linear(x, 3, 1)
    return y.reshape(1)
```

```python
Y_pred = []
for x in X:
    x = x.reshape(-1)
    y_pred = linear_network(x)
    Y_pred.append(y_pred)
plt.plot(Y_pred)
```
